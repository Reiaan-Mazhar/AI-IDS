# Transformer base model

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, random_split
import numpy as np

X_tensor = torch.tensor(np.array(X_train), dtype=torch.float32)
y_tensor = torch.tensor(np.array(y_train), dtype=torch.long)

dataset = TensorDataset(X_tensor, y_tensor)
train_loader = DataLoader(dataset, batch_size=64, shuffle=True)
```

```python
X_tensor = torch.tensor(np.array(X_test), dtype=torch.float32)
y_tensor = torch.tensor(np.array(y_test), dtype=torch.long)

dataset = TensorDataset(X_tensor, y_tensor)
val_loader = DataLoader(dataset, batch_size=64, shuffle=True)


class TabTransformer(nn.Module):
    def __init__(self, num_features=15, num_classes=2, embed_dim=64, n_heads=4, n_layers=2, ff_dim=128, dropout=0.1):
        super(TabTransformer, self).__init__()
        
        # Project each numeric feature to an embedding vector
        self.feature_embedding = nn.Linear(1, embed_dim)  # project each scalar to embedding
        
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embed_dim, nhead=n_heads, dim_feedforward=ff_dim, dropout=dropout, batch_first=True
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=n_layers)
        
        # Classifier head
        self.classifier = nn.Sequential(
            nn.Linear(num_features * embed_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        # x shape: (batch_size, num_features)
        x = x.unsqueeze(-1)  # (batch_size, num_features, 1)
        x = self.feature_embedding(x)  # (batch_size, num_features, embed_dim)
        x = self.transformer_encoder(x)  # (batch_size, num_features, embed_dim)
        x = x.flatten(start_dim=1)  # flatten for classifier
        out = self.classifier(x)
        return out

# Initialize model
model = TabTransformer(num_features=15, num_classes=2)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

epochs = 20

for epoch in range(epochs):
    model.train()
    total_loss = 0
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    # Validation
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for X_val, y_val in val_loader:
            preds = model(X_val)
            predicted = torch.argmax(preds, dim=1)
            total += y_val.size(0)
            correct += (predicted == y_val).sum().item()
    val_acc = correct / total
    print(f"Epoch [{epoch+1}/{epochs}]  Loss: {total_loss/len(train_loader):.4f}  Val Acc: {val_acc:.4f}")

torch.save(model.state_dict(), "tabtransformer_nslkdd_model.pth")
print("Transformer-based nslkdd model saved!")
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```
